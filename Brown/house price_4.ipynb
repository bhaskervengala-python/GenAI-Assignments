```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
```

```python
#sample data
area = np.array([ 1200, 1800, 2200,2350, 2500])
bedrooms = np.array([1, 2, 3,3.5, 4 ])
price = np.array([ 180000, 240000, 310000,340000, 360000])

# Combine features
X = np.column_stack((area, bedrooms))
y = price

# Train regression model
model = LinearRegression()
model.fit(X, y)

# Create mesh grid for regression plane
area_range = np.linspace(area.min(), area.max(), 20)
bedroom_range = np.linspace(bedrooms.min(), bedrooms.max(), 20)

A, B = np.meshgrid(area_range, bedroom_range)

# Predict prices for grid
Z = model.predict(np.column_stack((A.ravel(), B.ravel())))
Z = Z.reshape(A.shape)

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter points (actual data)
ax.scatter(area, bedrooms, price)

# Regression plane
ax.plot_surface(A, B, Z, alpha=0.5)

# Labels
ax.set_xlabel("Area (sq ft)")
ax.set_ylabel("Number of Bedrooms")
ax.set_zlabel("Price ($)")

plt.title("3D Regression Plane for House Prices")
plt.show()
```
